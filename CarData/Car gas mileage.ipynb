### 항목에 따른 자동차 연비 예측

##### 데이터 준비 및 탐색

```python
import numpy as np
import pandas as pd

data_df = pd.read_csv('C:/DataMining/CarData/auto-mpg.csv', header = 0, engine = 'python')
```

```python
print('데이터셋 크기: ', data_df.shape)
data_df.head()
```

```python
#피처 중에서 car_name, origin, horsepower는 분석에 사용하지 않으므로 제거 후 확인
data_df = data_df.drop(['car_name', 'origin', 'horsepower'], axis = 1, inplace = False)

data_df.head()
```

```python
print('데이터셋 크기: ', data_df.shape) #데이터셋 형태 확인
```

```python
data_df.info() #데이터셋 정보 확인
```

##### 분석 모델 구축, 결과 분석 및 시각화

1. 선형 회귀 분석 모델 구축하기

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
```

```python
#X, Y 분할하기
#mpg 피처를 회귀식의 종속 변수 Y로 설정하고, mpg를 제외한 나머지 피처를 독립 변수 X로 설정

Y = data_df['mpg']
X = data_df.drop(['mpg'], axis = 1, inplace = False)
```

```python
#훈련용 데이터와 평가용 데이터 분할하기
#데이터를 7:3 비율 test_size=0.3로 분할하여 train_test_split( ) 학습 데이터와 평가 데이터로 설정
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.3, random_state = 0)
```

```python
#선형 회귀 분석 : 모델 생성
lr = LinearRegression()
```

```python
#선형 회귀 분석 : 모델 훈련
lr.fit(X_train, Y_train)
```

```python
#선형 회귀 분석 : 평가 데이터에 대한 예측 수행 -> 예측 결과 Y_predict 구하기
Y_predict = lr.predict(X_test)
```

2. 평가 지표를 통해 선형 회귀 분석 모델을 평가하고 회귀 계수를 확인하여 자동차 연비에 끼치는 피처의 영향을 분석

```python
mse = mean_squared_error(Y_test, Y_predict)
rmse = np.sqrt(mse)

print('MSE : {0:.3f}, RMSE : {1:.3f}'.format(mse, rmse))
print('R^2(Variance score) : {0:.3f}'.format(r2_score(Y_test, Y_predict)))
```

```python
print('Y 절편 값: ', np.round(lr.intercept_, 2))
print('회귀 계수 값: ', np.round(lr.coef_, 2))
```

```python
coef = pd.Series(data = np.round(lr.coef_, 2), index = X.columns)
coef.sort_values(ascending = False)
```

회귀 모델 결과로 자동차 연비를 예측하는 회귀식

Ympg = -0.14Xcylinders + 0.01Xdisplacement - 0.01Xweight + 0.20Xacceleration + 0.76Xmodel_year – 17.55

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
fig, axs = plt.subplots(figsize = (16, 16), ncols = 3, nrows = 2)
x_features = ['model_year', 'acceleration', 'displacement', 'weight', 'cylinders']
plot_color = ['r', 'b', 'y', 'g', 'r']

for i, feature in enumerate(x_features):
     row = int(i/3)
     col = i%3
     sns.regplot(x = feature, y = 'mpg', data = data_df, ax = axs[row][col], color = plot_color[i])
```

3. 완성된 자동차 연비 예측 모델을 사용하여 임의의 데이터를 입력하면 연비를 예측할 수 있음

```python
print("연비를 예측하고 싶은 차의 정보를 입력해주세요.")
cylinders_1 = int(input("cylinders : "))
displacement_1 = int(input("displacement : "))
weight_1 = int(input("weight : "))
acceleration_1 = int(input("acceleration : "))
model_year_1 = int(input("model_year : "))
```

```python
mpg_predict = lr.predict([[cylinders_1, displacement_1, weight_1, acceleration_1 , model_year_1]])
```

```python
print("이 자동차의 예상 연비(MPG)는 %.2f입니다." %mpg_predict)
```
