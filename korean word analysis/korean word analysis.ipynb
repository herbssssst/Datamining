### 한글 뉴스 기사의 키워드 분석하기

1. 데이터 준비

```python
import json
import re
import matplotlib
import matplotlib.pyplot as plt
 
from konlpy.tag import Okt #Okt: 한글 품사 태깅을 위한 모듈
from collections import Counter

from matplotlib import font_manager, rc
from wordcloud import WordCloud
```

```python
#json 파일을 읽어 data 객체에 저장. 한글이 깨지지 않도록 utf-8 형식으로 인코딩
inputFileName = 'C:/DataMining/textData/etnews.kr_facebook_2016-01-01_2018-08-01_4차 산업혁명'

data = json.loads(open(inputFileName+'.json', 'r', encoding = 'utf8').read())
data
```

##### message 키의 데이터에서 품사가 명사인 단어만 추출

```python
message = ''

for item in data:
    if 'message' in item.keys():
        #[^\w]는 문자, 숫자, 밑줄이 아닌 모든 문자
        message = message + re.sub(r'[^\w]', ' ', item['message']) +''
message
```

```python
#명사만 추출하여 저장
nlp = Okt()

#analyze( text ) 텍스트를 다양한 형태로 반환, morphs( text ) 텍스트에서 형태소를 반환
#nouns( text ) 텍스트에서 명사를 반환, pos( text ) 텍스트에서 품사 정보를 부착하여 반환
message_N = nlp.nouns(message)
message_N
```

2. 데이터 탐색 및 분석 모델 구축

##### 명사가 저장된 message_N에 있는 단어 탐색

```python
# Counter() 함수를 사용하여 단어별 출현 횟수를 계산
count = Counter(message_N)
count 
```

```python
word_count = dict() 

#출현 횟수가 많은 상위 80개의 단어 중에서 길이가 1보다 큰 것만 word_count 딕셔너리에 저장하면서 출력하여 확인
for tag, counts in count.most_common(80): 
    if(len(str(tag))>1): 
        word_count[tag] = counts
        print("%s : %d" % (tag, counts))
```

##### 히스토그램을 통한 단어 빈도 탐색

```python
#히스토그램에 레이블을 한글로 표시하기 위해 한글 폰트인 맑은고딕체malgun.ttf를 설정matplotlib.rc( ) 
font_path = "c:/Windows/fonts/malgun.ttf"

font_name = font_manager.FontProperties(fname = font_path).get_name()
matplotlib.rc('font', family = font_name)
```

```python
#영문 단어 빈도 탐색과 같음
plt.figure(figsize = (12, 5))

plt.xlabel('키워드')
plt.ylabel('빈도수')

plt.grid(True)

sorted_Keys = sorted(word_count, key = word_count.get, reverse = True)
sorted_Values = sorted(word_count.values(), reverse = True)

plt.bar(range(len(word_count)), sorted_Values, align = 'center')
plt.xticks(range(len(word_count)), list(sorted_Keys), rotation = 75)

plt.show()
```

3. 결과 시각화

```python
wc = WordCloud(font_path, background_color = 'ivory', width = 800, height = 600)

cloud = wc.generate_from_frequencies(word_count)

plt.figure(figsize = (8, 8))
plt.imshow(cloud)
plt.axis('off')
plt.show()
```
